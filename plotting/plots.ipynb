```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.widgets import LassoSelector
```

```python
data = pd.read_csv('tanafull.csv')
```

```python
steps = 1831
interval = 1
step = np.linspace(0, interval, steps)
altitude = data.altitude
f_altitude = data.filtered_s
velocity = data.filtered_v
acceleration=data.filtered_a
```

```python
#plotting altitude
%matplotlib inline

plt.plot(step, altitude, '-', color='black')
plt.legend = ['Altitude']
plt.title = 'Altitude Plot'
plt.xlabel = 'Time (s)'
plt.ylabel = 'Altitude (m)'
plt.grid(True)
```

```python
#get maximum value in altitude
print('Maximum altitude: ', altitude.max(), 'm')
```

```python
# point with maximum altitude
max_altitude = altitude.idxmax()
max_altitude
```

```python
#plotting f_altitude
%matplotlib inline
plt.legend = ['Filtered Altitude']  
plt.title = 'Filtered Altitude Plot'
plt.xlabel = 'Time (s)'
plt.ylabel = 'Altitude (m)'
plt.grid(True)
plt.plot(step, f_altitude, '-', color='black')

```

```python
#maximum value in f_altitude
print('Maximum filtered altitude: ', f_altitude.max(), 'm')
```

```python
#point of maximum altitude
max_f_altitude = f_altitude.idxmax()
max_f_altitude
```

```python
#plotting velocity
plt.plot(step, velocity, '-', color='blue')

plt.legend = ['Velocity']
plt.title = 'Velocity Plot'
plt.xlabel = 'Time (s)'
plt.ylabel = 'Velocity (m/s)'
plt.grid(True)
```

```python
#point with zero at velocity
zero_velocity = velocity.idxmin()
zero_velocity
```

```python
#plotting acceleration
plt.plot(step, acceleration, '-', color='red')
plt.legend = ['Acceleration']
plt.title = 'Acceleration Plot'
plt.xlabel = 'Time (s)'
plt.ylabel = 'Acceleration (m/s^2)'
plt.grid(True)
```

```python
#maximum acceleration
print('Maximum acceleration: ', acceleration.max(), 'm/s^2')
```

```python
#point with maximum acceleration
max_acceleration = acceleration.idxmax()
max_acceleration
```
